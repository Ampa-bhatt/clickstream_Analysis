 **1.1 Definition**
The analysis on sequence of click and interaction that a user makes while navigating through a website on a digital platform.

**1.2 Use of Click Stream:**
Clicking link buttons and images as well as the path they take to purchase. The Clickstream analysis provides valuable insights into user behaviour and can help organisations understand how users are engaging with their online content. By analysing clickstream data, businesses can gain a deeper understanding of user preferences, interests, navigation patterns, and potential pain points. This information can then be used to make informed decisions regarding website design, content placement, user experience enhancements, and even marketing strategies.
1.  	**Page Views**: The number of times a particular page is viewed.
2.  	**Session**: The amount of time a user spends on the website during a single visit.
3.  	**Bounce Rate**: The percentage of visitors who navigate away from the site after viewing only one page.
4.  	**Conversion Paths**: The sequence of pages and interactions that lead to a desired action, such as making a purchase or filling out a form.
5.  	**Popular Pages:/Most Visited Pages** Identifying which pages are most frequently visited by users.
6.  	**Exit Pages**: The last pages users visit before leaving the website.
7.  	**Navigation Patterns**: How users move from one page to another and the paths they take.
8.  	**Referral Sources**: Where users are coming from before landing on the website (e.g., search engines, social media, other websites).

###4.0 **Summary of dataset**###
________________________________________________________________________________


1. **Client**: This column represents the IP address of the user or device that made the HTTP request to the website.

2. **Datetime**: This column records the datetime when the clickstream event occurred.

3. **Method**: It includes the request method (e.g., GET).

4. **Request**: This column contains the details of the HTTP request made by the user or device. It includes the URL path (e.g., /filter/27|13%20%D9%85%DA%AF%D8%A7%D9%...).

5. **Status**: This column indicates the HTTP response status code that the server returned in response to the request. For example, a "200" status code typically indicates a successful response.

6. **Size**: This column represents the size of the HTTP response in terms of bytes.

7. **Referer**: This column contains the address or URL of the website or specific page that was accessed.

8. **User_agent**: This column provides details about the user's browser, such as its user-agent string, which can include information about the browser type, version, and platform.

From this dataset, we can perform various types of clickstream analysis to gain insights into user behavior, website performance, and more:

- **Path Analysis**: By analyzing the sequence of URLs visited, you can identify common navigation paths users take through the website.

- **Referral Analysis**: You can determine where the traffic is coming from by examining the "Site_address" column and segment users based on referral sources.

- **Page Views and Engagement**: Calculate metrics like the number of page views, average response size, and engagement time to understand how users interact with the site.

- **Browser and Device Analysis**: Utilize the "Browser_details" column to analyze the distribution of browsers and devices used by visitors.

- **Time-based Patterns**: Analyze user behavior over different dates and times to identify patterns in site usage.

####4.1 **Status codes**####

**200**: Success - The request was successful, and the server has returned the requested data. (9579824 occurrences)

**304**: Not Modified - Indicates that the resource has not been modified since the last request, so it's still cached by the client. (340228 occurrences)

**302**: Found (Temporary Redirect) - The requested resource has been temporarily moved to a different URL. (199835 occurrences)

**404**: Not Found - The requested resource could not be found on the server. (105011 occurrences)

**301:** Moved Permanently - The requested resource has been permanently moved to a different URL. (67552 occurrences)

**499:** Client Closed Request - The client closed the connection before the server could respond. (50852 occurrences)

**500**: Internal Server Error - The server encountered an internal error while processing the request. (14266 occurrences)

**403**: Forbidden - The client does not have permission to access the requested resource. (5634 occurrences)

**502:** Bad Gateway - Indicates that a server acting as a gateway or proxy received an invalid response from an upstream server. (798 occurrences)

401: Unauthorized - Authentication is required, and the provided credentials are not valid. (323 occurrences)

**400**: Bad Request - The request is malformed or invalid. (318 occurrences)

**408**: Request Timeout - The client's request took too long to complete, and the server timed out waiting for it. (112 occurrences)

504: Gateway Timeout - The server acting as a gateway or proxy did not receive a timely response from an upstream server. (103 occurrences)

**405**: Method Not Allowed - The HTTP method used in the request is not allowed for the requested resource. (6 occurrences)

**206:** Partial Content - Indicates that the server is returning a partial representation of the resource. (The count is missing; it's usually used for ranged requests where only a portion of the resource is requested.)

##### 7.1 **Breaking sessions**#####
________________________________________________________________________________

It is recommended to set the session timeout to 10 minutes. This is a standard default setting that works well for most websites. That's why we chose the threshold to be 10 minutes.
[link](https://learn.microsoft.com/en-us/previous-versions/iis/6.0-sdk/ms525473(v=vs.90))

````python
df["time_diff"]=df.groupby("user_id")['timestamp'].diff().fillna(pd.Timedelta(0))
session_duration = 0
session_count = 1
current_user_id = None
sessions = []

for index, row in df.iterrows():
    if current_user_id != row['user_id']:
        current_user_id = row['user_id']
        session_duration = 0
        session_count = 1

    session_duration += row['time_diff'].total_seconds()

    if row['time_diff'].total_seconds() >600:  # check time
        session_count += 1
        session_duration = 0

    sessions.append(session_count)

df=df.drop(['time_diff'], axis=1)
df['session'] = sessions
````

Sequence Mining Algorithms:
**Apriori Algorithm:**

This algorithm is commonly used for market basket analysis, but it can also be adapted to analyse sequences of actions in clickstream data. It helps identify frequent sequences of actions that users take.

**GSP (Generalised Sequential Pattern) Algorithm:**

Like Apriori. GSP focuses on finding frequent sequences of actions in clickstreams. It is useful for understanding common user behaviour patterns.
**Markov Models:**

**First-order Markov Models**:

 These models assume that a user's next action depends only on their current action. They are useful for analysing the probability of transitions between pages or states.

**Higher-order Markov Models**: These models consider a user's current and previous actions to predict their next action. They can capture more complex behaviour patterns.
**PageRank Algorithm:**

Originally developed for ranking web pages, PageRank can be applied to clickstream data to identify important pages or actions in the user journey. It helps highlight pages that are frequently visited or serve as entry points.

**Clustering Algorithms:**

K-Means Clustering: By clustering users based on their clickstream behaviour, you can identify groups of users who follow similar patterns.

**DBSCAN (Density-Based Spatial Clustering of Applications with Noise):**

Useful for identifying groups of users who follow dense clickstream paths, helping you uncover specific segments.

Recommender Systems:
Collaborative Filtering: Based on users' historical clickstream data, you can recommend pages or products that similar users have interacted with.

Graph Algorithms:

**Shortest Path Algorithms:**
As mentioned earlier, these algorithms help analyse the shortest paths users take through the website, highlighting common navigation routes and areas for improvement.

**Time-Series Analysis:**

**Generalised Sequential Pattern**
The Algorithm applicable to sequential mining. It has importance to frequent sequence of events in context of click Stream Analysis. GSP can be applied to track user sequence of action clicks, page views and other in clickstream dataset.

1.      Defining Support Threshold: It is the minimum occurrence count, which is also called the support Threshold.
2.      Generating Candidate Sequences: This algorithm begins by identifying single events of the candidate, which are also called frequent sequences. It synthesises new candidate sequences and then gradually expands the length of sequence gradually.
3.      Counting Support: The dataset is scanned to count the occurrences of candidate sequences. The support count of a sequence is the number of times it appears in the dataset
4.      Pruning infrequent Sequence: Candidate sequences with support counts below the defined threshold are pruned, as they are not considered frequent. This helps reduce the number of sequences that need to be processed
5.      Generating New Candidates: The remaining candidate sequences are used to generate longer sequences by adding a new event to the end. These new sequences are then counted for support
6. Iterating and Expanding Sequences: The process of generating candidates, counting support, and pruning continues in an iterative manner until no more frequent sequences can be generated
7. Final Output: The algorithm returns a list of frequent sequences that meet the support threshold. users take when interacting with a website. For example, it could reveal that a frequent sequence of actions is "Home Page" → "Product Page" → "Add to Cart" → "Checkout." This insight can be valuable for optimising the user experience, understanding user preferences, and improving the design of the website's navigation flow.

**Markov Model (Markov Chain):**

1.State Visibility: In a Markov Chain, the state transitions are directly observable. This means that you can see the current state and the transition to the next state explicitly. The state transitions are determined solely by the current state, and the future states are independent of past states, given the current state.
2.Modeling Principle: Markov Chains are used to model situations where the probability of transitioning to a new state depends only on the current state. They are often used to model memoryless processes, where the future behaviour depends only on the present.
3.Applications: Markov Chains are commonly used in various fields, such as finance, weather prediction, game theory, and more. They are suitable for situations where events are influenced only by the immediate past.

**Hidden Markov Model (HMM):**

 1.State Visibility: In an HMM, the states are not directly observable; they are hidden. Instead, you observe a sequence of outputs (observations or emissions) that are associated with each hidden state.
2.Modeling Principle: HMMs are used to model situations where you have underlying states that influence the observable outputs, but you cannot directly observe the states. The observable outputs provide probabilistic information about the hidden states.
3.Applications: HMMs are widely used in various fields, including speech recognition, natural language processing, bioinformatics, and more. They are suitable for modeling scenarios where you want to infer the hidden states based on observed outputs.
 Key Differences

1.Observations: In a Markov Chain, you directly observe the states. In an HMM, you observe outputs associated with hidden states.

2.Information Source: Markov Chains depend on the current state to determine future states. HMMs depend on both the hidden states and observable outputs to make inferences.

3.Applications: Markov Chains are suitable for memoryless processes where the future is influenced solely by the current state. HMMs are used for modelling scenarios where underlying hidden states influence the observed outputs.

4.Complexity: HMMs tend to be more complex to work with because of the added layer of hidden states and observable outputs.

 In summary, the main distinction between a Markov Chain and a Hidden Markov Model is the presence of observable outputs associated with hidden states in the latter. Markov Chains model transitions between directly observable states, while HMMs handle scenarios where you observe outputs influenced by hidden states.
