**Summary of dataset**

---


1. **Client**: This column represents the IP address of the user or device that made the HTTP request to the website.

2. **Datetime**: This column records the datetime when the clickstream event occurred.

3. **Method**: It includes the request method (e.g., GET).

4. **Request**: This column contains the details of the HTTP request made by the user or device. It includes the URL path (e.g., /filter/27|13%20%D9%85%DA%AF%D8%A7%D9%...).

5. **Status**: This column indicates the HTTP response status code that the server returned in response to the request. For example, a "200" status code typically indicates a successful response.

6. **Size**: This column represents the size of the HTTP response in terms of bytes.

7. **Referer**: This column contains the address or URL of the website or specific page that was accessed.

8. **User_agent**: This column provides details about the user's browser, such as its user-agent string, which can include information about the browser type, version, and platform.

From this dataset, we can perform various types of clickstream analysis to gain insights into user behavior, website performance, and more:

- **Path Analysis**: By analyzing the sequence of URLs visited, you can identify common navigation paths users take through the website.

- **Referral Analysis**: You can determine where the traffic is coming from by examining the "Site_address" column and segment users based on referral sources.

- **Page Views and Engagement**: Calculate metrics like the number of page views, average response size, and engagement time to understand how users interact with the site.

- **Browser and Device Analysis**: Utilize the "Browser_details" column to analyze the distribution of browsers and devices used by visitors.

- **Time-based Patterns**: Analyze user behavior over different dates and times to identify patterns in site usage.

####**5.1.1 Breaking Sessions**####

###Session Standard###

As per standard practice, session duration can vary from industry to industry. Typically, it is around 10, 20, or 30 minutes. Most e-commerce companies follow a 20-minute session, while in some cases, it is 30 minutes. Banking service industries have a lower threshold of 5 or 2 minutes for transactional safety.

In our case, we follow the practice that if a user does not have any activity within 30 minutes of the threshold on a particular page, according to the timestamp, the present session ends. The next timestamp is considered to be the first page of the next session. Whenever the time spent on any page is greater than or equal to 30 minutes, it is considered as the last page of the preceding session. Hence, that page is the session exit page.

It is not possible to find the time spent on the exit page because the time spent on each page is estimated by subtracting the subsequent descending timestamp. In the pandas dataframe, there are some issues related to the estimation of time differences. This is because whenever we find the time spent on a particular page, the time difference defaults to giving a value of (n+1)th - nth timestamp for the nth page

###Methodology###





1)We can not take less than 30 minutes became on the basis of this. If we can not detect the user the user behaviour correctly we will lose information.


2) In this case the user do not need any prompt.

[link](https://learn.microsoft.com/en-us/previous-versions/iis/6.0-sdk/ms525473(v=vs.90)))

```python
df2["time_diff"]=df2.groupby("user_id")['timestamp'].diff().fillna(pd.Timedelta(0))
session_duration = 0
session_count = 1
current_user_id = None
sessions = []

for index, row in df2.iterrows():
    if current_user_id != row['user_id']:
        current_user_id = row['user_id']
        session_duration = 0
        session_count = 1

    session_duration += row['time_diff'].total_seconds()

    if row['time_diff'].total_seconds() >1800:  # check time
        session_count += 1
        session_duration = 0

    sessions.append(session_count)

df2=df2.drop(['time_diff'], axis=1)
df2['session'] = sessions
```

Z score= x-u/sigma  this is the formula we have taken.
x=data points,u=mean,sigma is the satandard deviation.if the zscore is greater than 3 we will consider that data point as an outlier.

```python
from scipy import stats
def remove_outliers_zscore(data2, column_name, threshold=3):
    z_scores = np.abs(stats.zscore(data2[column_name]))
    filtered_data2 = data2[(z_scores < threshold)]
    return filtered_data2

# Example data
data2 = pd.DataFrame({
    'Column1': s1['Time_spent_each_page'],
    'Column2': s1['Clicks']
})

# Choose the column on which you want to calculate Z score
column_name = 'Column1'

# Remove outliers using the function
filtered_data2 = remove_outliers_zscore(data2, column_name,threshold=1)

# Print the original and filtered data
print("Original data:")
print(data2)

print("\nFiltered data:")
print(filtered_data2)
```

The function takes two parameters: seq (the input sequence) and n_steps (the number of steps in each input sequence).
It initializes two empty lists, X and y, to store input and output sequences, respectively.
It iterates through the input sequence (seq) and creates input-output pairs of length n_steps.
If the last index (lastIndex) goes beyond the length of the input sequence, the loop breaks.
For each iteration, it creates an input sequence (seq_X) from the current index to lastIndex - 1 and the corresponding output value (seq_y) at index lastIndex.
These input and output sequences are appended to the X and y lists.
Finally, the function converts X and y into NumPy arrays and returns them.

```python
import numpy as np
data=df['page_id'].tolist()
def splitSequence(seq, n_steps):

    #Declare X and y as empty list
    X = []
    y = []

    for i in range(len(seq)):
        #get the last index
        lastIndex = i + n_steps

        #if lastIndex is greater than length of sequence then break
        if lastIndex > len(seq) - 1:
            break

        #Create input and output sequence
        seq_X, seq_y = seq[i:lastIndex], seq[lastIndex]

        #append seq_X, seq_y in X and y list
        X.append(seq_X)
        y.append(seq_y)
        pass
    #Convert X and y into numpy array
    X = np.array(X)
    y = np.array(y)

    return X,y

    pass
```

## 11.0 Levenshtein Distance

levenshtein_distance(seq1, seq2) is a Python function that calculates the Levenshtein distance between two input sequences seq1 and seq2.


The function initializes a dynamic programming (DP) table dp of dimensions (m+1) x (n+1), where m and n are the lengths of seq1 and seq2, respectively. This table will be used to store the intermediate results of the Levenshtein distance calculation.


The first two for loops initialize the DP table to account for the case where one of the sequences is empty. Specifically, dp[i][0] represents the cost of converting the first i characters of seq1 into an empty string, and dp[0][j] represents the cost of converting an empty string into the first j characters of seq2.


The nested for loops fill in the DP table. For each pair of characters in seq1 and seq2, it calculates the cost of various edit operations (insertion, deletion, and substitution) and selects the minimum cost, updating the DP table accordingly.


Finally, the function returns dp[m][n], which represents the Levenshtein distance between seq1 and seq2.

```python
def levenshtein_distance(seq1, seq2):
    m, n = len(seq1), len(seq2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m + 1):
        dp[i][0] = i

    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            cost = 0 if seq1[i - 1] == seq2[j - 1] else 1
            dp[i][j] = min(
                dp[i - 1][j] + 1,  # Insertion
                dp[i][j - 1] + 1,  # Deletion
                dp[i - 1][j - 1] + cost  # Substitution
            )

    return dp[m][n]

# Example usage:
seq1 =[1097, 1105, 1158]
seq2 = [3727, 3758, 3823]
distance = levenshtein_distance(seq1, seq2)
print("Levenshtein Distance:", distance)
```
